```python
# Imports
import numpy as np
import keras
from keras.datasets import imdb
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
from keras.preprocessing.text import Tokenizer
import matplotlib.pyplot as plt
from keras import optimizers
from fnn_helper import PlotLosses
%matplotlib inline

np.random.seed(42)
```

```python
# Loading the data (it's preloaded in Keras)
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=1000)

print(x_train.shape)
print(x_test.shape)
```

```python
print(x_train[0])
print(y_train[0])
```

```python
# Turning the output into vector mode, each of length 1000
tokenizer = Tokenizer(num_words=1000)
x_train = tokenizer.sequences_to_matrix(x_train, mode='binary')
x_test = tokenizer.sequences_to_matrix(x_test, mode='binary')
print(x_train.shape)
print(x_test.shape)
```

```python
x_train[0]
```

```python
# One-hot encoding the output
num_classes = 2
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
print(y_train.shape)
print(y_test.shape)
```

```python
# Entrenar el modelo
# Modifique batch_size y epochs si lo desea
hist = model.fit(x_train, y_train,
          batch_size=32,
          epochs=10,
          validation_data=(x_test, y_test), 
          verbose=2)
```

```python
# Evaluar el modelo
score = model.evaluate(x_test, y_test, verbose=0)
print("accuracy: ", score[1])
```

```python
# Building the model architecture with one layer of length 100
model = Sequential()
model.add(Dense(512, activation='relu', input_dim=1000))
model.add(Dense(num_classes, activation='softmax'))
model.summary()
sgd = optimizers.SGD()
# Compiling the model using categorical_crossentropy loss, and rmsprop optimizer.
model.compile(loss='categorical_crossentropy',
              optimizer=sgd,
              metrics=['accuracy'])
```

```python
# Running and evaluating the model
plot_losses = PlotLosses(plot_interval=1, evaluate_interval=100, x_val=x_test, y_val_categorical=y_test)
hist = model.fit(x_train, y_train,
          batch_size=32,
          epochs=50,
          validation_data=(x_test, y_test), 
          verbose=1, callbacks=[plot_losses])
```
